```python
df['Touch Screen'] = df['ScreenResolution'].str.contains('Touchscreen',na =False).map({True:'Yes',False:'No'})
```

```python
df['IPS'] = df['ScreenResolution'].str.contains('IPS',na=False).map({True:'Yes',False:'No'})
```

```python
df['Retina'] = df['ScreenResolution'].str.contains('Retina',na=False).map({True:'Yes',False:'No'})
```

Extracting Width , Height from Column


```python
df[['Width','Height']] = df['ScreenResolution'].str.extract(r'(\d+)x(\d+)').astype(float)
```

```python
res_map = {
    (1366.0, 768.0): 'HD / HD+',
    (1440.0, 900.0): 'HD / HD+',
    (1600.0, 900.0): 'HD / HD+',
    (1920.0, 1080.0): 'Full HD (FHD)',
    (1920.0, 1200.0): 'Full HD (FHD)',
    (2560.0, 1440.0): 'Quad HD (QHD)',
    (2256.0, 1504.0): 'Quad HD+ (QHD+)',
    (2304.0, 1440.0): 'Quad HD+ (QHD+)',
    (2560.0, 1600.0): 'Quad HD+ (QHD+)',
    (2736.0, 1824.0): 'Quad HD+ (QHD+)',
    (2880.0, 1800.0): 'Quad HD+ (QHD+)',
    (3200.0, 1800.0): 'Quad HD+ (QHD+)',
    (2400.0, 1600.0): 'Quad HD+ (QHD+)',
    (3840.0, 2160.0): '4K Ultra HD (UHD)'
}
```

```python
df['Maximum Display Resolution'] = pd.Series(list(zip(df['Width'], df['Height'],))).map(res_map)
```

```python
df['Maximum Display Resolution'] =df['Maximum Display Resolution'].fillna('Not Available')
```

```python
print(df['Maximum Display Resolution'].unique())
```

```python
df.drop(columns=['ScreenResolution'],inplace=True)
```

```python
df['Memory'] = df['Memory'].str.replace('?','')
```

```python
df['Memory'] = df['Memory'].replace('',np.nan)
```

```python
df['Menory'] = df['Memory'].dropna()
```

```python
df['Memory'] =df['Memory'].astype(str)
```

```python
Temp = df['Memory'].str.split('+',n=1,expand=True)
```

```python
df['Primary Storage'] = Temp[0].str.strip()
```

```python
df['Secondary Storage'] = Temp[1].str.strip().fillna('Not Available')
```

```python
df['Secondary Storage'].iloc[2]
```

```python
storage_type='SSD|HDD|Flash|Storage|Hybrid'
```

```python
df['Secondary Storage Type'] = df['Secondary Storage'].str.extract(f'({storage_type})').fillna('Not Available')
```

```python
df['Clock Speed'] =df['Cpu'].str.extract(r'(\d+(?:\.\d+)?GHz)').fillna('Not Available')
```
